## Data Analysis and Manipulation

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'mr':[2,3,4,5],
                    'days':[34,67,89,23]},
                  index = [2001,2009,2003,2020])
df1
```

```python
df2 = pd.DataFrame({'n_p':[3,6,5,9],
                   'n_h':[20,56,34,90]},
                  index = [2001, 2020,2034,2056])
df2
```

```python
df1.join(df2)
```

```python
df2.join(df1)
```

```python
df3 = pd.DataFrame({'a':[80,90,60,70],
                   'b':[2,1,2,3]},
                   index = [2001,2002,2003,2004])
df3
```

```python
df4 = pd.DataFrame({'a':[80,69,60,70],
                   'b':[2,1,2,3]},
                   index = [2001,2002,2003,2004])
df4
```

```python
df3.merge(df4)
```

```python
cont = pd.concat([df1,df2])
cont
```
